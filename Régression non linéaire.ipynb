```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import pca
from sklearn import linear_model
from mpl_toolkits.mplot3d import axes3d
```

```python
def f(x):
    return x*np.cos(x)+np.random.normal(size=500)*2

x=np.linspace(0,10,500)
rng=np.random.RandomState(0)
rng.shuffle(x)
x=np.sort(x[:])
y=f(x)
plt.plot(x,y,'o', color='black',markersize=2)
plt.axis('off')
plt.show()
```

```python
x_augmented = np.array([x, x**2, x**3,x**4,x**5]).T

regr = linear_model.LinearRegression()
regr.fit(x_augmented, y)
regr.score(x_augmented, y)
```

```python
plt.plot(x,y,'o', color='black',markersize=2)
plt.plot(x,regr.predict(x_augmented),'-', color='red',markersize=2)
plt.axis('off')
plt.show()
```
